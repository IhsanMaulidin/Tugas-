Import Library

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading DataSet

```python
url='/content/drive/MyDrive/Colab Notebooks/data0-SZL.csv'
df=pd.read_csv(url)
```

```python
df.head()
```

Perhitungan Statistika

```python
df.describe(include='all')
```

```python
df.skew()
```

```python
mean = np.mean(df)

# Menghitung energi (momen kedua)
energy = np.mean((df - mean) ** 2)

print("Energi:", energy)
```

```python
# Membuat subplot
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
# Menentukan kolom untuk plot
columns_to_plot = df.columns[:6]

# Plotting
for i, ax in enumerate(axes.flatten()):
    if i < len(columns_to_plot):
        df[columns_to_plot[i]].plot(ax=ax, kind='hist', title=columns_to_plot[i])
    else:
        ax.axis('off')  # Menghilangkan subplot yang tidak digunakan

plt.tight_layout()
plt.show()
```

```python
means = np.mean(df, axis=0)
stds = np.std(df, axis=0)

# Plot
plt.figure(figsize=(10, 6))
for i in range(0, 160, 10):
    plt.scatter(means[i:i+10], stds[i:i+10], label=f'Columns {i}-{i+9}')

plt.xlabel('Mean')
plt.ylabel('Standard Deviation')
plt.title('Scatter Plot: Mean vs Standard Deviation for every 10 columns')
plt.legend()
plt.grid(True)
plt.show()
```

```python
skewness = np.nanmean((df - np.nanmean(df, axis=0))**3 / np.nanstd(df, axis=0)**3, axis=0)
median = np.nanmedian(df, axis=0)

# Plot
plt.figure(figsize=(10, 6))
for i in range(0, 160, 10):
    plt.scatter(skewness[i:i+10], median[i:i+10], label=f'Columns {i}-{i+9}')

plt.xlabel('Skewness')
plt.ylabel('Median')
plt.title('Scatter Plot: Skewness vs Median for every 10 columns')
plt.legend()
plt.grid(True)
plt.show()
```
